```python
import xarray as xr
import numpy as np
```

```python
case_ctrl = 'fates_4x5_spmode_thetapsiitests_control_7616179e89_825579d0'
case_mod1 = 'fates_4x5_spmode_thetapsiitests_nearzero_7616179e89_8b8065c8'
case_mod2 = 'fates_4x5_spmode_thetapsiitests_pt9_7616179e89_54262997'

fin_ctrl = xr.open_dataset('/global/homes/c/cdkoven/scratch/e3sm_scratch/pm-cpu/'+case_ctrl+'/run/'+case_ctrl+'.60years.nc')
fin_mod1 = xr.open_dataset('/global/homes/c/cdkoven/scratch/e3sm_scratch/pm-cpu/'+case_mod1+'/run/'+case_mod1+'.60years.nc')
fin_mod2 = xr.open_dataset('/global/homes/c/cdkoven/scratch/e3sm_scratch/pm-cpu/'+case_mod2+'/run/'+case_mod2+'.60years.nc')
```

### Map of mean GPP (kg C / m2 / yr) in control case

```python
(fin_ctrl['FATES_GPP']*86400*365).isel(time=slice(240,None)).mean(dim='time').plot()
```

### Map of mean GPP anomalies (kg C / m2 / yr) relative to control case. 
### First is theta_cii = near zero, second is theta_cii = 0.9

```python
((fin_mod1['FATES_GPP'] - fin_ctrl['FATES_GPP'])*86400*365).isel(time=slice(240,None)).mean(dim='time').plot()
```

```python
((fin_mod2['FATES_GPP'] - fin_ctrl['FATES_GPP'])*86400*365).isel(time=slice(240,None)).mean(dim='time').plot()
```

### Map of relative GPP anomalies (fraction) relative to control case. 
### First is theta_cii = near zero, second is theta_cii = 0.9

```python
((fin_mod1['FATES_GPP'] - fin_ctrl['FATES_GPP']).isel(time=slice(240,None)).mean(dim='time') / fin_ctrl['FATES_GPP'].isel(time=slice(240,None)).mean(dim='time')).plot(levels = (-0.3,-0.25,-0.2,-0.15,-0.1,-0.05,0,0.05,0.1,0.15,0.2,0.25,0.3))

```

```python
((fin_mod2['FATES_GPP'] - fin_ctrl['FATES_GPP']).isel(time=slice(240,None)).mean(dim='time') / fin_ctrl['FATES_GPP'].isel(time=slice(240,None)).mean(dim='time')).plot(levels = (-0.3,-0.25,-0.2,-0.15,-0.1,-0.05,0,0.05,0.1,0.15,0.2,0.25,0.3))
```
